```python
# Sum of two values

# Given an array of integers and a value, determine if there are any two integers in the array whose sum is equal to the given value.

#this solution works by computing what number we need from the target value and the value we're currently looking at in the array


#so we can generate a random array
import numpy as np
from sys import exit

#ask user for array length
arr_input = input("How large should the array be?: ")

#make sure its an integer
try:
    arr_size = int(arr_input)
except ValueError:
    print("that's not an int you mook!")
    raise

#create array of chosen length
numpy_arr = np.random.randint(1000, size= arr_size)
#convert numpy to python list/array so we can use pop...
arr = list(numpy_arr)

#print the array (useful for actually verifying if the program is correct...)
print("For the array: {}".format(arr))

#get the target value from the user
target_input = input("What number should I look for?: ")

#make sure target is an integer
try:
    target_value = int(target_input)
except ValueError:
    print("that's not an int you mook!")
    raise


#to track if found for output message below
found = False

#for every element in the array
for i in range(0, len(arr)):
    
    #get the target value minus the first number (the indexed number) and 
    # remove it from the list to prevent the possibility of duplicate ie [3,2,1] looking for 6 resuling in 0,0
    temp = target_value - arr.pop(0)
    
    #check if our temp value is in the array (i.e, our current index has a counterpart that adds to the target value)
    if temp in arr:
        #i is where the arr.pop happens, the second value is +1 to account for the pop and plus i to account for previous pops
        print("indexes {} and {} add up to {}".format(i, i + arr.index(temp) + 1, target_value))
        found = True

#if not found, inform the user
if found == False:
    print("I couldn't find any two indexes in the array that add up to {}".format(target_value))
    
```

```python
# Move zeros to the left

# Move all zeros to the left of an array while maintaining its order.

#this solution works by having two scrolling indexs beginning at the end of the array that move left until they hit the front of the array
# they begin by being side by size i.e. index and index+1, but for every zero they encounter the gap between them is increased.
# the last index should always be pointing to a zero

arr_input = input("How large should the array be?: ")

#make sure its an integer
try:
    arr_size = int(arr_input)
except ValueError:
    print("that's not an int you mook!")
    raise

    
range_input = input("What should the range of values be? (0 to ___?): ")

#make sure its an integer
try:
    range_value = int(range_input)+1
except ValueError:
    print("that's not an int you mook!")
    raise
    

arr = np.random.randint(range_value, size= arr_size)

index_end = len(arr)-1
index_front = index_end-1

#while we haven't touched the first element in the array
while index_front != -1:
    #switch cases!
    
    #if the end of our window rests on a zero, we may hunt for a number to replace it with...
    if arr[index_end] == 0: 
        if arr[index_front] != 0: #if we have found a number to replace it with...
            
            #swap them!
            temp = arr[index_end]
            arr[index_end] = arr[index_front]
            arr[index_front] = temp
            
            # and advance the end of our window since it no longer points to a zero
            index_end-=1
            
        elif arr[index_front] == 0: #if we haven't found a number yet...
            index_front-=1 #increase the size of the window!

    #if the end index is NOT a zero... we want it to be
    elif arr[index_end] != 0: 
        
        # advance the end index until it hits a zero OR it is next to the front index
        
        if index_end-1 == index_front: #arr indexes side by side
            #shift them both up by 1!
            index_end-=1
            index_front-=1
        else:
            #advance the end of the window to find a zero!
            index_end-=1
    print(arr)
    
    
```

```python
# Delete node with given key

# You are given the head of a linked list and a key. You have to delete the node that contains this given key.

import numpy as np


class Node:

    #here we create an init function for inception of a node. This includes a previous pointer and a next pointer
    def __init__(self, value):
        self.value = value
        self.next = None
        
    #here we create a representation function for displaying the node if called upon
    def __repr__(self):
        return self.value

class linked_list:
    
    #another init for the head of the list. This includes a counter for elements in the list and a tail pointer for finding the end quickly
    def __init__(self):
        self.head = None
     
    #here we have a representation function for displaying the linked list itself, it works by traversing the list and storing the value of each node in an array
    #and then joining the array together separated by arrows ( -> )
    def __repr__(self):
        cur = self.head
        nodes = []

        while cur != None:
            nodes.append(str(cur.value))
            cur = cur.next

        nodes.append("None")

        return " -> ".join(nodes)
    
 #insert the node at the head of the linked list
    def insert_at_head(self, value):
        
        new_node = Node(value)

        if self.head == None: #if no head exists
            #set the head to be the new node
            self.head = new_node

        else: #if there is a current head
            #point the new head's next to be the current head
            new_node.next = self.head

            #reassign the list's head pointer to be the new head
            self.head = new_node
        

    def delete_node_with_value(self, target_value):

        #if the head is being deleted
        if self.head.value == target_value:
            #set the value after head to be the new head
            self.head = self.head.next
                
        else:
            #to track the previous node
            just_before = self.head
            cur = self.head.next
            
            #to escape the loop
            deleted = False
        
            #while not at the end of the list and node hasn't been deleted
            while cur != None and deleted != True:
                
                #yay! we found our target node!
                if cur.value == target_value:
                    #set the previous node's next pointer to be the one after the target node
                    #i.e. cut it out of the list!
                    just_before.next = cur.next
                    
                    #let's get out of the loop!
                    deleted = True
                else:
                    #advance our positions!
                    just_before = cur
                    cur = cur.next
                

                         
#create an array of random values    
arr = np.random.randint(100, size= 20)

#create an empty list
the_list = linked_list()

#add every element in the array to the list
for x in arr:
    the_list.insert_at_head(x)
    
done = False

print(the_list)

while done != True:
    #get a value to delete from the user
    to_delete_input = (input("Enter a number to delete (if done, enter -1): "))
    try:
        to_delete = int(to_delete_input)
        # see if the user wants to kill the cell
        if to_delete == -1:
            done = True

        else: #delete the value
            the_list.delete_node_with_value(to_delete)
            print(the_list)
    except ValueError:
        print("that's not an int you mook!")
        
        
```

```python
# Copy linked list with arbitrary pointer

# You are given a linked list where the node has two pointers. The first is the regular ‘next’ pointer. 
# The second pointer is called ‘arbitrary_pointer’ and it can point to any node in the linked list.

# Your job is to write code to make a deep copy of the given linked list. Here, deep copy means that any
# operations on the original list (inserting, modifying and removing) should not affect the copied list.

from random import randint
from random import shuffle
import numpy as np

class Node:

    #here we create an init function for inception of a node. This includes a previous pointer and a next pointer
    def __init__(self, value):
        self.value = value
        self.next = None
        self.arbitrary_pointer = None
        
    #here we create a representation function for displaying the node if called upon
    def __repr__(self):
        if self.arbitrary_pointer == None:
            return str(self.value) + "-None"
        else:
            return str(self.value) +"-"+ str(self.arbitrary_pointer.value)

    
class arbitrary_linked_list:
    
    def __init__(self):
        self.head = None

    def __repr__(self):
        cur = self.head
        
        while cur != None:
            
            print(cur, end = "") 
            print("->", end = "")

            cur = cur.next

        print("none")
        return ""
       
    
    def insert_at_head(self, value):
        
        new_node = Node(value)


        if self.head == None: #if no head exists
            #set the head to be the new node
            self.head = new_node

        else: #if there is a current head
            #point the new head's next to be the current head
            new_node.next = self.head

            #reassign the list's head pointer to be the new head
            self.head = new_node
    
    def insert_at_tail(self, value):
        
        new_node = Node(value)
        
        if self.head == None: #if no head exists
            #set the head to be the new node
            self.head = new_node
            
        else:
            cur = self.head
            #while not at the end, advance our position
            while cur.next != None:
                
                cur = cur.next
            
            #now we should be at the end, so just insert the node
            cur.next = new_node
    
    def delete_node_with_value(self, target_value):

        #if the head is being deleted
        if self.head.value == target_value:
            #set the value after head to be the new head
            self.head = self.head.next
                
        else:
            #to track the previous node
            just_before = self.head
            cur = self.head.next
            
            #to escape the loop
            deleted = False
        
            #while not at the end of the list and node hasn't been deleted
            while cur != None and deleted != True:
                
                #yay! we found our target node!
                if cur.value == target_value:
                    #set the previous node's next pointer to be the one after the target node
                    #i.e. cut it out of the list!
                    just_before.next = cur.next
                    
                    #let's get out of the loop!
                    deleted = True
                else:
                    #advance our positions!
                    just_before = cur
                    cur = cur.next
                    
    def set_arbitrary_arbitrarily(self): #see what i did there? ;)
        cur = self.head
        list_values = []
        
        #while not at the end of the list
        while cur != None:
            #let's get an array of the values in the list..
            list_values.append(cur.value)
            cur = cur.next
        
        #now let's do that again but pare off values from the list to randomly point to!
        shuffle(list_values)
        index = 0
        cur = self.head
        while cur != None:
            value_to_look_for = list_values[index]
            
            #find the node with that value!
            arbitrary_node = self.head
            while arbitrary_node.value != value_to_look_for:
                arbitrary_node = arbitrary_node.next
            
            #now that we have it, point the current node at it *arbitrarily*
            cur.arbitrary_pointer = arbitrary_node
            
            
            index+=1
            cur = cur.next
        
    
    #let's make a function to solve this problem!
    def create_copy(self):
        copy_list = arbitrary_linked_list()
        
        cur = self.head
        #first let's copy the regular list, without the arbitrary pointer
        while cur != None:
            new_node_val = cur.value
            
            copy_list.insert_at_tail(new_node_val)
            
            cur = cur.next
        
        #now let's copy the arbitrary pointer!
        cur = self.head
        cur_copy = copy_list.head
        while cur != None:
            
            
            #get the value of the arbitrary node!
            arbitrary_val = cur.arbitrary_pointer.value
            
            #find that node in our list!
            target_node = copy_list.head
            while target_node.value != arbitrary_val:
                target_node = target_node.next
            
            #now that we have the node, let's assign our current node to point 'arbitrarily' at it!
            cur_copy.arbitrary_pointer = target_node
            
            #now advance down the lists!
            cur = cur.next
            cur_copy = cur_copy.next
        
        return copy_list

#create an array of random values    
arr = np.random.randint(100, size= 20)

#create an empty list
the_list = arbitrary_linked_list()

#add every element in the array to the list
for x in arr:
    the_list.insert_at_head(x)

the_list.set_arbitrary_arbitrarily()

copied = the_list.create_copy()   
    
done = False

print("The original list:")
print(the_list)
print("The copied list:")
print(copied)


while done != True:
    #get a value to delete from the user
    to_delete_input = (input("Enter a number to delete (if done, enter -1): "))
    try:
        to_delete = int(to_delete_input)
        # see if the user wants to kill the cell
        if to_delete == -1:
            done = True

        else: #delete the value
            the_list.delete_node_with_value(to_delete)
            print("The original list :")
            print(the_list)
            print("The copied list :")
            print(copied)
    except ValueError:
        print("that's not an int you mook!")
```

```python
# Mirror binary trees

# Given the root node of a binary tree, swap the 'left' and 'right' children for each node.

#the basework of this is shamelessly copied from my BST implementatino found under python/datastructures/binary_search_tree.py

import numpy as np

class Node:
    def __init__(self, value):

        #save that value
        self.value = value
        #and then two pointers for it's children
        self.childL = None
        self.childR = None

    def __repr__(self):
        return str(self.value)
    
class BST:

    def __init__(self, root = None):

        self.root = root

        #to keep track of how tall the tree is, may be useful when displaying it
        self.height = 0

    def __repr__(self):
        self.print_tree

    # this function inserts a node into the array using recursion, parent is automatically set to none 
    # on first call and then used to determine place in the tree as the function progresses
    def insert(self, node, parent = None):

        #if new tree
        if self.root == None:
            #this new node is now the root
            self.root = node

        
        else:
            #else check if this is first call
            if parent == None:
            #if first call, assign parent to root
                parent = self.root

            #find out where the node should go
            if node.value < parent.value:
                #if it goes to the left, does it have a new node to branch from?
                if parent.childL != None:
                    #if it does, call insert on that node
                    self.insert(node, parent.childL)
                else:
                    #if it doesn't, insert the node here!
                    parent.childL = node

            #rinse and repeat for the right side...
            if node.value >= parent.value:
                #if it goes to the left, does it have a new node to branch from?
                if parent.childR != None:
                    #if it does, call insert on that node
                    self.insert(node, parent.childR)
                else:
                    #if it doesn't, insert the node here!
                    parent.childR = node

    def print_tree(self, height = 0, parent = None):
        #check if tree even exists
        if self.root == None:
            print("This tree is empty")
        else:
            #if no parent, we are at the start of the tree
            if parent  == None:
                #assign parent to root
                parent = self.root

            #print the current node
            print("\t"*height + str(parent.value))

            #if node has a left child
            if parent.childL != None:
                print("\t"*(height+1) + "-L-")
                self.print_tree(height + 1, parent = parent.childL)

            if parent.childR != None:
                print("\t"*(height+1) + "-R-")
                self.print_tree(height + 1, parent = parent.childR)
                
    def mirror(self, node = None):
        if node == None:
 
            node = self.root
        
        #flip this node's children
        
        temp = node.childR
        node.childR = node.childL
        node.childL = temp
        
        #see if children are nodes themselves.. and if so, flip 'em!
        
        if node.childR != None:
            self.mirror(node.childR)
        if node.childL !=None:
            self.mirror(node.childL)

#create an empty tree            
the_tree = BST()

#create an array of random values    
arr = np.random.randint(100, size= 20)

#insert them to the tree
for x in arr:
    the_tree.insert(Node(x))
    
the_tree.print_tree()
the_tree.mirror()
the_tree.print_tree()
```
